```python
import json
import re
from collections import defaultdict

# --- Load properties from JSON file ---
with open("property_data.json", "r", encoding="utf-8") as f:
    properties_data = json.load(f)

def get_numeric_value(value):
    """Extract integer from strings like '800 sqft', '10 years', etc."""
    if not value:
        return 0
    match = re.search(r"\d+", str(value))
    return int(match.group()) if match else 0

def filter_properties(user_input, data, field="size"):
    """
    Filters and sorts properties by Size, Carpet Area, or Age.
    field can be 'size', 'carpet', or 'age'.
    """
    user_input = user_input.lower().strip()
    filtered_properties = []

    # Decide which field to use
    if field == "size":
        data_field = "Size_In_Sqft"
    elif field == "carpet":
        data_field = "Carpet_Area_Sqft"
    elif field == "age":
        data_field = "Property_Age"
    else:
        data_field = "Size_In_Sqft"

    # --- Check if user entered a property ID ---
    match = re.match(r"property_(\d+)", user_input, re.IGNORECASE)
    if match:
        pid = f"Property_{match.group(1)}"
        for prop in data:
            if str(prop.get("property_id", "")).lower() == pid.lower():
                return {"Specific Property": [prop]}
        return {}

    # --- Handle "below X" ---
    match = re.search(r"below\s+(\d+)", user_input)
    if match:
        limit = int(match.group(1))
        filtered_properties = [p for p in data if get_numeric_value(p.get(data_field)) < limit]

    # --- Handle "above X" ---
    match = re.search(r"above\s+(\d+)", user_input)
    if match:
        limit = int(match.group(1))
        filtered_properties = [p for p in data if get_numeric_value(p.get(data_field)) > limit]

    # --- Handle "between X and Y" ---
    match = re.search(r"between\s+(\d+)\s+and\s+(\d+)", user_input)
    if match:
        low, high = int(match.group(1)), int(match.group(2))
        filtered_properties = [p for p in data if low <= get_numeric_value(p.get(data_field)) <= high]

    # --- Handle "exactly X" or just number ---
    match = re.search(r"(\d+)", user_input)
    if match and not filtered_properties:
        value = int(match.group(1))
        filtered_properties = [p for p in data if get_numeric_value(p.get(data_field)) == value]

    if not filtered_properties:
        return {}

    # --- Group by property type ---
    grouped_properties = defaultdict(list)
    for prop in filtered_properties:
        property_type = prop.get("Room_Details", {}).get("Type", "Other/Unspecified Type")
        grouped_properties[property_type].append(prop)

    # --- Sort each group ---
    for prop_type in grouped_properties:
        grouped_properties[prop_type].sort(key=lambda x: get_numeric_value(x.get(data_field)))

    return dict(grouped_properties)

def display_property(prop, full=False):
    """Pretty print property details"""
    if full:
        print("\n--- Full Property Details ---")
        for key, value in prop.items():
            print(f"{key}: {value}")
    else:
        print(
            f"- ID: {prop.get('property_id','N/A')} | Rent: ₹{prop.get('Rent_Price','N/A')} | "
            f"Size: {prop.get('Size_In_Sqft','Unknown')} sqft | "
            f"Carpet Area: {prop.get('Carpet_Area_Sqft','Unknown')} sqft | "
            f"Age: {prop.get('Property_Age','Unknown')}"
        )

def chatbot_response():
    print("Welcome to the Property Finder Chatbot!")
    print("I can help you find properties in Nagpur by Size, Carpet Area, Property Age, or Property ID.")

    while True:
        print("\nDo you want to search by:")
        print("1. Size")
        print("2. Carpet Area")
        print("3. Property Age")
        print("4. Property ID (e.g., Property_992)")
        user_choice = input("(Type the number or 'exit' to quit): ").strip()

        if user_choice.lower() == "exit":
            print("Thank you for using the Property Finder Chatbot. Goodbye!")
            break

        if user_choice not in ["1", "2", "3", "4"]:
            print("❌ Invalid choice. Please enter 1, 2, 3 or 4.")
            continue

        if user_choice == "4":
            user_input = input("\nEnter Property ID (e.g., Property_992): ").strip()
            results = filter_properties(user_input, properties_data, "size")
            if results:
                for _, props in results.items():
                    for prop in props:
                        display_property(prop, full=True)
            else:
                print(f"No property found with ID '{user_input}'.")
            continue

        search_map = {"1": "size", "2": "carpet", "3": "age"}
        search_type = search_map[user_choice]
        user_input = input(
            f"\nEnter {search_type} query (e.g., '10 years', 'below 5', 'between 2 and 8'): "
        ).strip()

        results = filter_properties(user_input, properties_data, search_type)

        if results:
            print(f"\nResults for '{user_input}' ({search_type}):")
            for prop_type, props in results.items():
                print(f"\n--- {prop_type} ---")
                for prop in props:
                    display_property(prop)
        else:
            print(f"No properties found for '{user_input}'. Please try another query.")


# --- Run the Chatbot ---
if __name__ == "__main__":
    chatbot_response()
```

def match_numeric(value, user_input):
    value = int(value)
    if user_input.startswith("below"):

        return value < num
    elif user_input.startswith("above"):
        num = int(re.findall(r'\d+', user_input)[0])
        return value > num
    elif user_input.startswith("between"):
        nums = [int(x) for x in re.findall(r'\d+', user_input)]

    else:
        num = int(re.findall(r'\d+', user_input)[0])

```python
import json
from collections import defaultdict

# --- Load properties from JSON file ---
try:
    with open("property_data.json", "r") as f:
        properties_data = json.load(f)
except FileNotFoundError:
    print("Error: 'property_data.json' not found. Please ensure the file exists.")
    properties_data = []
except json.JSONDecodeError:
    print("Error: Could not decode 'property_data.json'. Please check its format.")
    properties_data = []

# --- Helper functions for normalization ---
def normalize_area_name(area_name):
    """Normalize an area name for consistent comparison."""
    return str(area_name).replace(" ", "").lower().strip()

def normalize_zone_name(zone_name):
    """Normalize a zone name for consistent comparison."""
    return str(zone_name).replace(" ", "").lower().strip()

def normalize_facility_name(facility_name):
    """Normalize a facility name for consistent comparison."""
    return str(facility_name).replace(" ", "_").lower().strip()

def normalize_amenity_name(name):
    """Normalize a nearby amenity name for consistent comparison."""
    return str(name).replace(" ", "_").lower().strip()

def normalize_room_name(room_name):
    """Normalize a room type name for consistent comparison (e.g., '1 BHK' -> '1bhk')."""
    return str(room_name).replace(" ", "").lower().strip()

def normalize_property_type_name(prop_type):
    """Normalize a property type name for consistent comparison (e.g., 'Flat' -> 'flat')."""
    return str(prop_type).lower().strip()

def normalize_key(key):
    """Normalize a user-provided key for mapping to search fields."""
    return key.lower().replace(" ", "_").replace("_type", "")

# --- Dynamically get all unique values from the dataset ---
ALL_FACILITIES = []
for p in properties_data:
    if "Facilities" in p and isinstance(p["Facilities"], dict):
        ALL_FACILITIES = sorted(list(p["Facilities"].keys()))
        break

ALL_NEARBY_AMENITIES = []
for p in properties_data:
    if "Nearby_Amenities" in p and isinstance(p["Nearby_Amenities"], dict):
        ALL_NEARBY_AMENITIES = sorted(list(p["Nearby_Amenities"].keys()))
        break

ALL_SEARCH_KEYS = [
    "size", "carpet", "age", "brokerage", "id", "amenities", "furnishing", "security",
    "rent", "area", "zone", "bedrooms", "bathrooms", "balcony", "floor_no", "total_floors",
    "maintenance", "recommended_for", "water_supply", "society_type", "road_connectivity",
    "facilities", "nearby_amenities", "room_type", "property_type"
]

def filter_properties(user_input, field, data):
    normalized_user_input = user_input
    if field == "area":
        normalized_user_input = normalize_area_name(user_input)
    elif field == "zone":
        normalized_user_input = normalize_zone_name(user_input)
    elif field == "room_type":
        normalized_user_input = normalize_room_name(user_input)
    elif field == "property_type":
        normalized_user_input = normalize_property_type_name(user_input)
    else:
        if field not in ["size", "carpet", "age", "security", "rent", "bedrooms", "bathrooms", "balcony", "floor_no", "total_floors", "road_connectivity"]:
            normalized_user_input = str(user_input).lower().strip()

    filtered_properties = []

    data_field_map = {
        "size": "Size_In_Sqft", "carpet": "Carpet_Area_Sqft", "age": "Age_Of_Property", "brokerage": "Brokerage",
        "furnishing": "Furnishing_Status", "amenities": "Amenities", "security": "Security_Deposite", "rent": "Rent_Price",
        "area": "Area", "zone": "Zone", "bedrooms": "Bedrooms", "bathrooms": "Bathrooms", "balcony": "Balcony",
        "floor_no": "Floor_No", "total_floors": "Total_floors_In_Building", "maintenance": "Maintenance_Charge",
        "recommended_for": "Recommended_For", "water_supply": "Water_Supply_Type", "society_type": "Society_Type",
        "road_connectivity": "Road_Connectivity", "facilities": "Facilities", "nearby_amenities": "Nearby_Amenities",
        "room_type": "Room_Details", "property_type": "Room_Details", "id": "property_id"
    }
    data_field = data_field_map.get(field)
    if not data_field:
        return {}

    if field in ["brokerage", "furnishing", "maintenance", "recommended_for", "water_supply", "society_type"]:
        filtered_properties = [p for p in data if str(p.get(data_field, "N/A")).lower() == normalized_user_input]
    elif field == "amenities":
        filtered_properties = [p for p in data if normalized_user_input in [a.lower().strip() for a in p.get("Amenities", [])]]
    elif field == "facilities":
        user_facilities = [normalize_facility_name(f) for f in user_input.split(',')]
        filtered_properties = [p for p in data if all(p.get("Facilities", {}).get(f, 0) == 1 for f in user_facilities)]
    elif field == "nearby_amenities":
        user_amenities = [normalize_amenity_name(f) for f in user_input.split(',')]
        filtered_properties = [p for p in data if all(p.get("Nearby_Amenities", {}).get(f, 0) == 1 for f in user_amenities)]
    elif field == "room_type":
        filtered_properties = [p for p in data if normalize_room_name(p.get("Room_Details", {}).get("Rooms", "")) == normalized_user_input]
    elif field == "property_type":
        filtered_properties = [p for p in data if normalize_property_type_name(p.get("Room_Details", {}).get("Type", "")) == normalized_user_input]
    elif field in ["area", "zone"]:
        filtered_properties = [p for p in data if (normalize_area_name(p.get("Area", "N/A")) if field == "area" else normalize_zone_name(p.get("Zone", "N/A"))) == normalized_user_input]
    elif field == "id":
        filtered_properties = [p for p in data if p.get(data_field) == user_input.strip()]
    else:
        try:
            if user_input.startswith("below"):
                value = float(user_input.split()[1])
                filtered_properties = [p for p in data if isinstance(p.get(data_field), (int, float)) and p[data_field] < value]
            elif user_input.startswith("above"):
                value = float(user_input.split()[1])
                filtered_properties = [p for p in data if isinstance(p.get(data_field), (int, float)) and p[data_field] > value]
            elif user_input.startswith("between"):
                parts = user_input.replace("between", "").replace("and", "").split()
                low, high = float(parts[0]), float(parts[1])
                filtered_properties = [p for p in data if isinstance(p.get(data_field), (int, float)) and low <= p[data_field] <= high]
            else:
                value = float("".join([c for c in user_input if c.isdigit() or c == "."]))
                filtered_properties = [p for p in data if isinstance(p.get(data_field), (int, float)) and p[data_field] == value]
        except (ValueError, IndexError):
            print(f"Invalid numeric input for {field}: '{user_input}'. Please enter a number or a range.")
            return []
        except Exception as e:
            print(f"An unexpected error occurred during numeric filtering: {e}")
            return []

    return filtered_properties

def format_property(prop):
    property_id = prop.get('property_id', 'N/A')
    rent_price = prop.get('Rent_Price', 'N/A')
    size = prop.get('Size_In_Sqft', 'Unknown')
    carpet_area = prop.get('Carpet_Area_Sqft', 'Unknown')
    security_deposit = prop.get('Security_Deposite', 'N/A')
    brokerage = prop.get('Brokerage', 'N/A')
    furnishing_status = prop.get('Furnishing_Status', 'N/A')
    amenities = ', '.join(prop.get('Amenities', []))
    age = prop.get('Age_Of_Property', 'Unknown')
    area = prop.get('Area', 'N/A')
    zone = prop.get('Zone', 'N/A')
    bedrooms = prop.get('Bedrooms', 'N/A')
    bathrooms = prop.get('Bathrooms', 'N/A')
    balcony = prop.get('Balcony', 'N/A')
    floor_no = prop.get('Floor_No', 'N/A')
    total_floors = prop.get('Total_floors_In_Building', 'N/A')
    maintenance = prop.get('Maintenance_Charge', 'N/A')
    recommended_for = prop.get('Recommended_For', 'N/A')
    water_supply = prop.get('Water_Supply_Type', 'N/A')
    society_type = prop.get('Society_Type', 'N/A')
    road_connectivity = prop.get('Road_Connectivity', 'N/A')
    facilities_list = [k.replace("_", " ").title() for k, v in prop.get("Facilities", {}).items() if v == 1]
    facilities = ', '.join(facilities_list) if facilities_list else 'None'
    nearby_amenities_list = [k.replace("_", " ").title() for k, v in prop.get("Nearby_Amenities", {}).items() if v == 1]
    nearby_amenities = ', '.join(nearby_amenities_list) if nearby_amenities_list else 'None'
    rooms = prop.get("Room_Details", {}).get("Rooms", "N/A")
    property_type = prop.get("Room_Details", {}).get("Type", "N/A")

    return (
        f"ID: {property_id} | Rent: ₹{rent_price} | Size: {size} sqft | Carpet Area: {carpet_area} sqft\n"
        f"Rooms: {rooms} | Property Type: {property_type} | Bedrooms: {bedrooms} | Bathrooms: {bathrooms}\n"
        f"Furnishing: {furnishing_status} | Security Deposit: ₹{security_deposit} | Brokerage: {brokerage}\n"
        f"Facilities: {facilities}\n"
        f"Nearby Amenities: {nearby_amenities}\n"
        f"Amenities: {amenities}\n"
        f"Floor: {floor_no}/{total_floors} | Maintenance: {maintenance} | Recommended For: {recommended_for}\n"
        f"Water Supply: {water_supply} | Society: {society_type} | Road Connectivity: {road_connectivity} km\n"
        f"Age: {age} years | Area: {area} | Zone: {zone}\n"
    )

def chatbot_response():
    print("Welcome to the Property Search Chatbot!")
    print("You can search for properties by combining multiple criteria.")
    print("Use the format: key=value + key=value + ...")
    print("For example: area=Manewada + rent=below 15000 + bedrooms=2")
    print("Type 'exit' to quit.")

    while True:
        combined_input = input("\nEnter your search query: ").strip()
        if combined_input.lower() == "exit":
            print("Goodbye!")
            break
        
        # Start with the full dataset
        filtered_properties = properties_data
        
        # Parse the combined input
        criteria = [c.strip() for c in combined_input.split('+')]
        
        valid_search = True
        for c in criteria:
            if '=' not in c:
                print(f"❌ Invalid format for criterion: '{c}'. Must be 'key=value'.")
                valid_search = False
                break
            
            key, value = c.split('=', 1)
            search_key = normalize_key(key)
            
            if search_key not in ALL_SEARCH_KEYS:
                print(f"❌ Invalid search key: '{key}'. Please use a valid key.")
                valid_search = False
                break
            
            # Filter the current results based on the new criterion
            filtered_properties = filter_properties(value, search_key, filtered_properties)
            
            if not filtered_properties:
                print("❌ No properties found matching the criteria so far. Resetting search.")
                valid_search = False
                break
        
        if not valid_search:
            continue
        
        # Group final results by property type
        grouped_results = defaultdict(list)
        for prop in filtered_properties:
            property_type = prop.get("Room_Details", {}).get("Type", "Other/Unspecified Type")
            grouped_results[property_type].append(prop)

        if not grouped_results:
            print("❌ No properties found matching your complete query.")
            continue
            
        print(f"\n✅ Found {len(filtered_properties)} properties matching your query.")
        for prop_type, props in grouped_results.items():
            print(f"\n--- 🏠 Property Type: {prop_type.title()} ({len(props)} results) ---")
            for prop in props:
                print(format_property(prop))

if __name__ == "__main__":
    chatbot_response()
```

```python
import json
from collections import defaultdict

# --- Load properties from JSON file ---
try:
    with open("property_data.json", "r") as f:
        properties_data = json.load(f)
except FileNotFoundError:
    print("Error: 'property_data.json' not found. Please ensure the file exists.")
    properties_data = []
except json.JSONDecodeError:
    print("Error: Could not decode 'property_data.json'. Please check its format.")
    properties_data = []

# --- Helper functions for normalization ---
def normalize_area_name(area_name):
    """Normalize an area name for consistent comparison."""
    return str(area_name).replace(" ", "").lower().strip()

def normalize_zone_name(zone_name):
    """Normalize a zone name for consistent comparison."""
    return str(zone_name).replace(" ", "").lower().strip()

def normalize_facility_name(facility_name):
    """Normalize a facility name for consistent comparison."""
    return str(facility_name).replace(" ", "_").lower().strip()

def normalize_amenity_name(name):
    """Normalize a nearby amenity name for consistent comparison."""
    return str(name).replace(" ", "_").lower().strip()

def normalize_room_name(room_name):
    """Normalize a room type name for consistent comparison (e.g., '1 BHK' -> '1bhk')."""
    return str(room_name).replace(" ", "").lower().strip()

def normalize_property_type_name(prop_type):
    """Normalize a property type name for consistent comparison (e.g., 'Flat' -> 'flat')."""
    return str(prop_type).lower().strip()

def normalize_key(key):
    """Normalize a user-provided search key for internal mapping."""
    # Convert common user inputs like "room type" to "room_type"
    key = key.lower().strip()
    if key == "room type":
        return "room_type"
    if key == "property type":
        return "property_type"
    if key == "floor number":
        return "floor_no"
    if key == "total floors":
        return "total_floors"
    if key == "maintenance charge":
        return "maintenance"
    if key == "recommended for":
        return "recommended_for"
    if key == "water supply type":
        return "water_supply"
    if key == "society type":
        return "society_type"
    if key == "road connectivity":
        return "road_connectivity"
    # For other keys, just replace spaces with underscores and make lowercase
    return key.replace(" ", "_")

# --- Dynamically get all unique values from the dataset ---
ALL_FACILITIES = []
for p in properties_data:
    if "Facilities" in p and isinstance(p["Facilities"], dict):
        ALL_FACILITIES = sorted(list(p["Facilities"].keys()))
        break

ALL_NEARBY_AMENITIES = []
for p in properties_data:
    if "Nearby_Amenities" in p and isinstance(p["Nearby_Amenities"], dict):
        ALL_NEARBY_AMENITIES = sorted(list(p["Nearby_Amenities"].keys()))
        break

# List of all valid search keys that the chatbot understands
ALL_SEARCH_KEYS = [
    "size", "carpet", "age", "brokerage", "id", "amenities", "furnishing", "security",
    "rent", "area", "zone", "bedrooms", "bathrooms", "balcony", "floor_no", "total_floors",
    "maintenance", "recommended_for", "water_supply", "society_type", "road_connectivity",
    "facilities", "nearby_amenities", "room_type", "property_type"
]

def filter_properties(user_input, field, data):
    """Filters a list of properties based on a single criterion."""
    normalized_user_input = user_input
    if field == "area":
        normalized_user_input = normalize_area_name(user_input)
    elif field == "zone":
        normalized_user_input = normalize_zone_name(user_input)
    elif field == "room_type":
        normalized_user_input = normalize_room_name(user_input)
    elif field == "property_type":
        normalized_user_input = normalize_property_type_name(user_input)
    # For other categorical fields, normalize the input directly
    elif field in ["brokerage", "furnishing", "maintenance", "recommended_for", "water_supply", "society_type", "amenities"]:
        normalized_user_input = str(user_input).lower().strip()
    # For numeric and ID fields, user_input is handled as is or parsed specifically

    filtered_properties = []

    # Map search field to the actual key in properties_data
    data_field_map = {
        "size": "Size_In_Sqft", "carpet": "Carpet_Area_Sqft", "age": "Age_Of_Property", "brokerage": "Brokerage",
        "furnishing": "Furnishing_Status", "amenities": "Amenities", "security": "Security_Deposite", "rent": "Rent_Price",
        "area": "Area", "zone": "Zone", "bedrooms": "Bedrooms", "bathrooms": "Bathrooms", "balcony": "Balcony",
        "floor_no": "Floor_No", "total_floors": "Total_floors_In_Building", "maintenance": "Maintenance_Charge",
        "recommended_for": "Recommended_For", "water_supply": "Water_Supply_Type", "society_type": "Society_Type",
        "road_connectivity": "Road_Connectivity", "facilities": "Facilities", "nearby_amenities": "Nearby_Amenities",
        "room_type": "Room_Details", "property_type": "Room_Details", "id": "property_id"
    }
    data_field = data_field_map.get(field)
    if not data_field:
        return [] # Should not happen with proper ALL_SEARCH_KEYS setup

    if field in ["brokerage", "furnishing", "maintenance", "recommended_for", "water_supply", "society_type"]:
        filtered_properties = [p for p in data if str(p.get(data_field, "N/A")).lower() == normalized_user_input]
    elif field == "amenities":
        # Check if the requested amenity is in the property's amenities list
        filtered_properties = [p for p in data if normalized_user_input in [a.lower().strip() for a in p.get("Amenities", [])]]
    elif field == "facilities":
        # Handle comma-separated list of facilities
        user_facilities = [normalize_facility_name(f) for f in user_input.split(',')]
        filtered_properties = [p for p in data if all(p.get("Facilities", {}).get(f, 0) == 1 for f in user_facilities)]
    elif field == "nearby_amenities":
        # Handle comma-separated list of nearby amenities
        user_amenities = [normalize_amenity_name(f) for f in user_input.split(',')]
        filtered_properties = [p for p in data if all(p.get("Nearby_Amenities", {}).get(f, 0) == 1 for f in user_amenities)]
    elif field == "room_type":
        filtered_properties = [p for p in data if normalize_room_name(p.get("Room_Details", {}).get("Rooms", "")) == normalized_user_input]
    elif field == "property_type":
        filtered_properties = [p for p in data if normalize_property_type_name(p.get("Room_Details", {}).get("Type", "")) == normalized_user_input]
    elif field == "area":
        filtered_properties = [p for p in data if normalize_area_name(p.get("Area", "N/A")) == normalized_user_input]
    elif field == "zone":
        filtered_properties = [p for p in data if normalize_zone_name(p.get("Zone", "N/A")) == normalized_user_input]
    elif field == "id":
        # Exact match for property_id
        filtered_properties = [p for p in data if str(p.get(data_field)) == user_input.strip()]
    else:
        # Numeric fields (size, carpet, age, security, rent, bedrooms, bathrooms, balcony, floor_no, total_floors, road_connectivity)
        try:
            if user_input.startswith("below"):
                value = float(user_input.split()[1])
                filtered_properties = [p for p in data if isinstance(p.get(data_field), (int, float)) and p[data_field] < value]
            elif user_input.startswith("above"):
                value = float(user_input.split()[1])
                filtered_properties = [p for p in data if isinstance(p.get(data_field), (int, float)) and p[data_field] > value]
            elif user_input.startswith("between"):
                parts = user_input.replace("between", "").replace("and", "").split()
                low, high = float(parts[0]), float(parts[1])
                filtered_properties = [p for p in data if isinstance(p.get(data_field), (int, float)) and low <= p[data_field] <= high]
            else:
                value = float("".join([c for c in user_input if c.isdigit() or c == "."]))
                filtered_properties = [p for p in data if isinstance(p.get(data_field), (int, float)) and p[data_field] == value]
        except (ValueError, IndexError):
            print(f"Error: Invalid numeric input for {field}: '{user_input}'. Please enter a number or a range.")
            return []
        except Exception as e:
            print(f"An unexpected error occurred during numeric filtering for {field}: {e}")
            return []

    return filtered_properties

def format_property(prop):
    """Formats a single property's details for display."""
    property_id = prop.get('property_id', 'N/A')
    rent_price = prop.get('Rent_Price', 'N/A')
    size = prop.get('Size_In_Sqft', 'Unknown')
    carpet_area = prop.get('Carpet_Area_Sqft', 'Unknown')
    security_deposit = prop.get('Security_Deposite', 'N/A')
    brokerage = prop.get('Brokerage', 'N/A')
    furnishing_status = prop.get('Furnishing_Status', 'N/A')
    amenities = ', '.join(prop.get('Amenities', []))
    age = prop.get('Age_Of_Property', 'Unknown')
    area = prop.get('Area', 'N/A')
    zone = prop.get('Zone', 'N/A')
    bedrooms = prop.get('Bedrooms', 'N/A')
    bathrooms = prop.get('Bathrooms', 'N/A')
    balcony = prop.get('Balcony', 'N/A')
    floor_no = prop.get('Floor_No', 'N/A')
    total_floors = prop.get('Total_floors_In_Building', 'N/A')
    maintenance = prop.get('Maintenance_Charge', 'N/A')
    recommended_for = prop.get('Recommended_For', 'N/A')
    water_supply = prop.get('Water_Supply_Type', 'N/A')
    society_type = prop.get('Society_Type', 'N/A')
    road_connectivity = prop.get('Road_Connectivity', 'N/A')
    facilities_list = [k.replace("_", " ").title() for k, v in prop.get("Facilities", {}).items() if v == 1]
    facilities = ', '.join(facilities_list) if facilities_list else 'None'
    nearby_amenities_list = [k.replace("_", " ").title() for k, v in prop.get("Nearby_Amenities", {}).items() if v == 1]
    nearby_amenities = ', '.join(nearby_amenities_list) if nearby_amenities_list else 'None'
    rooms = prop.get("Room_Details", {}).get("Rooms", "N/A")
    property_type = prop.get("Room_Details", {}).get("Type", "N/A")

    return (
        f"ID: {property_id} | Rent: ₹{rent_price} | Size: {size} sqft | Carpet Area: {carpet_area} sqft\n"
        f"Rooms: {rooms} | Property Type: {property_type} | Bedrooms: {bedrooms} | Bathrooms: {bathrooms}\n"
        f"Furnishing: {furnishing_status} | Security Deposit: ₹{security_deposit} | Brokerage: {brokerage}\n"
        f"Facilities: {facilities}\n"
        f"Nearby Amenities: {nearby_amenities}\n"
        f"Amenities: {amenities}\n"
        f"Floor: {floor_no}/{total_floors} | Maintenance: {maintenance} | Recommended For: {recommended_for}\n"
        f"Water Supply: {water_supply} | Society: {society_type} | Road Connectivity: {road_connectivity} km\n"
        f"Age: {age} years | Area: {area} | Zone: {zone}\n"
        f"----------------------------------------"
    )

def chatbot_response():
    print("Welcome to the Property Search Chatbot!")
    print("You can search for properties by combining multiple criteria using '+'.")
    print("Example: area=Manewada + rent=below 15000 + bedrooms=2")
    print("Available keys: " + ", ".join(ALL_SEARCH_KEYS))
    print("Type 'exit' to quit.")

    while True:
        combined_input = input("\nEnter your search query: ").strip()
        if combined_input.lower() == "exit":
            print("Goodbye!")
            break
        
        # Start with the full dataset
        current_filtered_properties = properties_data
        
        # Parse the combined input
        criteria_parts = [c.strip() for c in combined_input.split('+')]
        
        valid_query = True
        for criterion in criteria_parts:
            if '=' not in criterion:
                print(f"❌ Invalid format for criterion: '{criterion}'. Each criterion must be in 'key=value' format.")
                valid_query = False
                break
            
            key_raw, value = criterion.split('=', 1)
            search_key = normalize_key(key_raw)
            
            if search_key not in ALL_SEARCH_KEYS:
                print(f"❌ Invalid search key: '{key_raw}'. Please use one of the available keys.")
                valid_query = False
                break
            
            # Apply filter sequentially
            current_filtered_properties = filter_properties(value, search_key, current_filtered_properties)
            
            if not current_filtered_properties:
                print(f"❌ No properties found after filtering by '{key_raw}={value}'.")
                valid_query = False
                break
        
        if not valid_query:
            continue
        
        # Group final results by property type for organized output
        grouped_results = defaultdict(list)
        for prop in current_filtered_properties:
            property_type = prop.get("Room_Details", {}).get("Type", "Other/Unspecified Type")
            grouped_results[property_type].append(prop)

        if not grouped_results:
            print("❌ No properties found matching your complete query.")
            continue
            
        print(f"\n✅ Found {len(current_filtered_properties)} properties matching your query.")
        for prop_type, props in grouped_results.items():
            print(f"\n--- 🏠 Property Type: {prop_type.title()} ({len(props)} results) ---")
            for prop in props:
                print(format_property(prop))

if __name__ == "__main__":
    chatbot_response()
```

```python
import json
from collections import defaultdict

# --- Load properties from JSON file ---
try:
    with open("property_data.json", "r", encoding="utf-8") as f:
        properties_data = json.load(f)
except FileNotFoundError:
    print("Error: 'property_data.json' not found. Please ensure the file exists.")
    properties_data = []
except json.JSONDecodeError:
    print("Error: Could not decode 'property_data.json'. Please check its format.")
    properties_data = []

# --- Helper functions for normalization ---
def normalize_area_name(area_name):
    return str(area_name).replace(" ", "").lower().strip()

def normalize_zone_name(zone_name):
    return str(zone_name).replace(" ", "").lower().strip()

def normalize_facility_name(facility_name):
    return str(facility_name).replace(" ", "_").lower().strip()

def normalize_amenity_name(name):
    return str(name).replace(" ", "_").lower().strip()

def normalize_room_name(room_name):
    return str(room_name).replace(" ", "").lower().strip()

def normalize_property_type_name(prop_type):
    return str(prop_type).lower().strip()

# --- Dynamically get all unique values from the dataset ---
ALL_AREAS = sorted(list(set(normalize_area_name(p.get("Area", "N/A")) for p in properties_data)))
ALL_ZONES = sorted(list(set(normalize_zone_name(p.get("Zone", "N/A")) for p in properties_data)))

ALL_FACILITIES = []
for p in properties_data:
    if "Facilities" in p and isinstance(p["Facilities"], dict):
        ALL_FACILITIES = sorted(list(p["Facilities"].keys()))
        break

ALL_NEARBY_AMENITIES = []
for p in properties_data:
    if "Nearby_Amenities" in p and isinstance(p["Nearby_Amenities"], dict):
        ALL_NEARBY_AMENITIES = sorted(list(p["Nearby_Amenities"].keys()))
        break

ALL_ROOM_TYPES = sorted(list(set(normalize_room_name(p.get("Room_Details", {}).get("Rooms", "N/A")) for p in properties_data if p.get("Room_Details", {}).get("Rooms"))))
ALL_PROPERTY_TYPES = sorted(list(set(normalize_property_type_name(p.get("Room_Details", {}).get("Type", "N/A")) for p in properties_data if p.get("Room_Details", {}).get("Type"))))

# --- Filter function ---
def filter_properties(user_input, field, data):
    normalized_user_input = user_input
    if field == "area":
        normalized_user_input = normalize_area_name(user_input)
    elif field == "zone":
        normalized_user_input = normalize_zone_name(user_input)
    elif field == "room_type":
        normalized_user_input = normalize_room_name(user_input)
    elif field == "property_type":
        normalized_user_input = normalize_property_type_name(user_input)
    else:
        if field not in ["size", "carpet", "age", "security", "rent", "bedrooms", "bathrooms", "balcony", "floor_no", "total_floors", "road_connectivity"]:
            normalized_user_input = str(user_input).lower().strip()

    filtered_properties = []

    data_field_map = {
        "size": "Size_In_Sqft", "carpet": "Carpet_Area_Sqft", "age": "Age_Of_Property", "brokerage": "Brokerage",
        "furnishing": "Furnishing_Status", "amenities": "Amenities", "security": "Security_Deposite", "rent": "Rent_Price",
        "area": "Area", "zone": "Zone", "bedrooms": "Bedrooms", "bathrooms": "Bathrooms", "balcony": "Balcony",
        "floor_no": "Floor_No", "total_floors": "Total_floors_In_Building", "maintenance": "Maintenance_Charge",
        "recommended_for": "Recommended_For", "water_supply": "Water_Supply_Type", "society_type": "Society_Type",
        "road_connectivity": "Road_Connectivity", "facilities": "Facilities", "nearby_amenities": "Nearby_Amenities",
        "room_type": "Room_Details", "property_type": "Room_Details", "id": "property_id"
    }
    data_field = data_field_map.get(field)
    if not data_field:
        return []

    try:
        if field in ["brokerage", "furnishing", "maintenance", "recommended_for", "water_supply", "society_type"]:
            filtered_properties = [p for p in data if str(p.get(data_field, "N/A")).lower() == normalized_user_input]
        elif field == "amenities":
            filtered_properties = [p for p in data if normalized_user_input in [a.lower().strip() for a in p.get("Amenities", [])]]
        elif field == "facilities":
            user_facilities = [normalize_facility_name(f) for f in user_input.split(',')]
            filtered_properties = [p for p in data if all(p.get("Facilities", {}).get(f, 0) == 1 for f in user_facilities)]
        elif field == "nearby_amenities":
            user_amenities = [normalize_amenity_name(f) for f in user_input.split(',')]
            filtered_properties = [p for p in data if all(p.get("Nearby_Amenities", {}).get(f, 0) == 1 for f in user_amenities)]
        elif field == "room_type":
            filtered_properties = [p for p in data if normalize_room_name(p.get("Room_Details", {}).get("Rooms", "")) == normalized_user_input]
        elif field == "property_type":
            filtered_properties = [p for p in data if normalize_property_type_name(p.get("Room_Details", {}).get("Type", "")) == normalized_user_input]
        elif field == "area":
            filtered_properties = [p for p in data if normalize_area_name(p.get("Area", "N/A")) == normalized_user_input]
        elif field == "zone":
            filtered_properties = [p for p in data if normalize_zone_name(p.get("Zone", "N/A")) == normalized_user_input]
        elif field == "id":
            filtered_properties = [p for p in data if str(p.get(data_field)) == user_input.strip()]
        else:
            # Numeric fields
            if user_input.startswith("below"):
                value = float(user_input.split()[1])
                filtered_properties = [p for p in data if isinstance(p.get(data_field), (int, float)) and p[data_field] < value]
            elif user_input.startswith("above"):
                value = float(user_input.split()[1])
                filtered_properties = [p for p in data if isinstance(p.get(data_field), (int, float)) and p[data_field] > value]
            elif user_input.startswith("between"):
                parts = user_input.replace("between", "").replace("and", "").split()
                low, high = float(parts[0]), float(parts[1])
                filtered_properties = [p for p in data if isinstance(p.get(data_field), (int, float)) and low <= p[data_field] <= high]
            else:
                value = float("".join([c for c in user_input if c.isdigit() or c == "."]))
                filtered_properties = [p for p in data if isinstance(p.get(data_field), (int, float)) and p[data_field] == value]
    except Exception as e:
        print(f"Error during filtering '{field}': {e}")
        return []

    return filtered_properties

# --- Format property for display ---
def format_property(prop):
    property_id = prop.get('property_id', 'N/A')
    rent_price = prop.get('Rent_Price', 'N/A')
    size = prop.get('Size_In_Sqft', 'Unknown')
    carpet_area = prop.get('Carpet_Area_Sqft', 'Unknown')
    security_deposit = prop.get('Security_Deposite', 'N/A')
    brokerage = prop.get('Brokerage', 'N/A')
    furnishing_status = prop.get('Furnishing_Status', 'N/A')
    amenities = ', '.join(prop.get('Amenities', []))
    age = prop.get('Age_Of_Property', 'Unknown')
    area = prop.get('Area', 'N/A')
    zone = prop.get('Zone', 'N/A')
    bedrooms = prop.get('Bedrooms', 'N/A')
    bathrooms = prop.get('Bathrooms', 'N/A')
    balcony = prop.get('Balcony', 'N/A')
    floor_no = prop.get('Floor_No', 'N/A')
    total_floors = prop.get('Total_floors_In_Building', 'N/A')
    maintenance = prop.get('Maintenance_Charge', 'N/A')
    recommended_for = prop.get('Recommended_For', 'N/A')
    water_supply = prop.get('Water_Supply_Type', 'N/A')
    society_type = prop.get('Society_Type', 'N/A')
    road_connectivity = prop.get('Road_Connectivity', 'N/A')
    facilities_list = [k.replace("_", " ").title() for k, v in prop.get("Facilities", {}).items() if v == 1]
    facilities = ', '.join(facilities_list) if facilities_list else 'None'
    nearby_amenities_list = [k.replace("_", " ").title() for k, v in prop.get("Nearby_Amenities", {}).items() if v == 1]
    nearby_amenities = ', '.join(nearby_amenities_list) if nearby_amenities_list else 'None'
    rooms = prop.get("Room_Details", {}).get("Rooms", "N/A")
    property_type = prop.get("Room_Details", {}).get("Type", "N/A")

    return (
        f"ID: {property_id} | Rent: ₹{rent_price} | Size: {size} sqft | Carpet Area: {carpet_area} sqft\n"
        f"Rooms: {rooms} | Property Type: {property_type} | Bedrooms: {bedrooms} | Bathrooms: {bathrooms}\n"
        f"Furnishing: {furnishing_status} | Security Deposit: ₹{security_deposit} | Brokerage: {brokerage}\n"
        f"Facilities: {facilities}\n"
        f"Nearby Amenities: {nearby_amenities}\n"
        f"Amenities: {amenities}\n"
        f"Floor: {floor_no}/{total_floors} | Maintenance: {maintenance} | Recommended For: {recommended_for}\n"
        f"Water Supply: {water_supply} | Society: {society_type} | Road Connectivity: {road_connectivity} km\n"
        f"Age: {age} years | Area: {area} | Zone: {zone}\n"
        f"----------------------------------------"
    )

# --- Chatbot ---
def chatbot_response():
    print("Welcome to Property Search Chatbot!")
    print("\nSearch Options:")
    print(
        "1. Size (sqft)\n2. Carpet Area (sqft)\n3. Age of Property\n4. Brokerage (yes/no)\n5. Property ID (exact match)\n"
        "6. Amenities\n7. Furnishing Status\n8. Security Deposit\n9. Rent Price\n10. Area\n11. Zone\n"
        "12. Bedrooms\n13. Bathrooms\n14. Balcony\n15. Floor Number\n16. Total Floors\n17. Maintenance Charge\n"
        "18. Recommended For\n19. Water Supply Type\n20. Society Type\n21. Road Connectivity\n22. Facilities\n"
        "23. Nearby Amenities\n24. Room Type (e.g., 1 BHK)\n25. Property Type (e.g., Flat)\n"
    )
    print("💡 You can combine multiple search fields using '+'.")
    print("Example: 1+2+3 (Size + Carpet Area + Age of Property)")
    print("Type 'exit' to quit.")

    search_map = {
        "1": "size", "2": "carpet", "3": "age", "4": "brokerage", "5": "id", "6": "amenities", "7": "furnishing",
        "8": "security", "9": "rent", "10": "area", "11": "zone", "12": "bedrooms", "13": "bathrooms",
        "14": "balcony", "15": "floor_no", "16": "total_floors", "17": "maintenance", "18": "recommended_for",
        "19": "water_supply", "20": "society_type", "21": "road_connectivity", "22": "facilities", "23": "nearby_amenities",
        "24": "room_type", "25": "property_type"
    }

    while True:
        field_numbers = input("\nEnter the number(s) of search fields to combine (e.g., 1+2+3) or 'exit': ").strip().lower()
        if field_numbers == "exit":
            print("Goodbye!")
            break

        selected_fields = []
        is_valid_input = True
        for num in field_numbers.split('+'):
            num = num.strip()
            if num in search_map:
                selected_fields.append(search_map[num])
            else:
                print(f"❌ Invalid field number: '{num}'. Please use a valid number.")
                is_valid_input = False
                break
        if not is_valid_input:
            continue

        combined_filters = {}
        for field in selected_fields:
            if field == "id":
                combined_filters[field] = input("Enter Property ID: ").strip()
                break
            elif field == "facilities":
                print("\nAvailable Facilities:", ', '.join([f.replace("_", " ").title() for f in ALL_FACILITIES]))
                combined_filters[field] = input("Enter facilities (comma separated): ").strip()
            elif field == "nearby_amenities":
                print("\nAvailable Nearby Amenities:", ', '.join([f.replace("_", " ").title() for f in ALL_NEARBY_AMENITIES]))
                combined_filters[field] = input("Enter nearby amenities (comma separated): ").strip()
            elif field == "room_type":
                print("\nAvailable Room Types:", ', '.join([r.upper() for r in ALL_ROOM_TYPES]))
                combined_filters[field] = input("Enter Room Type: ").strip()
            elif field == "property_type":
                print("\nAvailable Property Types:", ', '.join([p.title() for p in ALL_PROPERTY_TYPES]))
                combined_filters[field] = input("Enter Property Type: ").strip()
            elif field == "area":
                print("\nAvailable Areas:", ', '.join([p.get("Area") for p in properties_data]))
                combined_filters[field] = input("Enter Area: ").strip()
            elif field == "zone":
                print("\nAvailable Zones:", ', '.join([p.get("Zone") for p in properties_data]))
                combined_filters[field] = input("Enter Zone: ").strip()
            else:
                combined_filters[field] = input(f"Enter {field} value: ").strip()

        # Special case: ID search
        if "id" in combined_filters:
            prop = next((p for p in properties_data if str(p.get("property_id")) == combined_filters["id"]), None)
            if prop:
                print("\n✅ Property Found:")
                print(format_property(prop))
            else:
                print("❌ Property not found.")
            continue

        # Apply combined filters sequentially
        results = properties_data
        for field, value in combined_filters.items():
            results = filter_properties(value, field, results)

        if not results:
            print("❌ No properties found matching your combined search.")
        else:
            print(f"\n✅ Found {len(results)} properties matching your query.")
            grouped_results = defaultdict(list)
            for prop in results:
                prop_type = prop.get("Room_Details", {}).get("Type", "Other/Unspecified Type")
                grouped_results[prop_type].append(prop)

            for prop_type, props in grouped_results.items():
                print(f"\n--- 🏠 Property Type: {prop_type.title()} ({len(props)} results) ---")
                for prop in props:
                    print(format_property(prop))

if __name__ == "__main__":
    chatbot_response()
```
