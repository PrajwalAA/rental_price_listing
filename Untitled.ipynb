```python
import json
import re
from collections import defaultdict

# --- Load properties from JSON file ---
try:
    with open("property_data.json", "r") as f:
        properties_data = json.load(f)
except FileNotFoundError:
    print("Error: 'property_data.json' not found. Please ensure the file exists.")
    properties_data = []
except json.JSONDecodeError:
    print("Error: Could not decode 'property_data.json'. Please check its format.")
    properties_data = []


# --- Normalization helpers ---
def normalize(text):
    return str(text).lower().strip().replace(" ", "_")

def get_numeric_value(value):
    if not value:
        return None
    match = re.search(r"\d+", str(value))
    return int(match.group()) if match else None


# --- Filtering logic ---
def filter_properties(user_input, data):
    user_input = user_input.lower()
    results = data

    # Rent filtering
    rent_match = re.search(r"rent\s*(below|under|less than|above|over|greater than)?\s*(\d+)", user_input)
    if rent_match:
        direction = rent_match.group(1)
        amount = int(rent_match.group(2))
        if direction in ["below", "under", "less than"]:
            results = [p for p in results if get_numeric_value(p.get("Rent_Price")) and get_numeric_value(p.get("Rent_Price")) < amount]
        elif direction in ["above", "over", "greater than"]:
            results = [p for p in results if get_numeric_value(p.get("Rent_Price")) and get_numeric_value(p.get("Rent_Price")) > amount]
        else:
            results = [p for p in results if get_numeric_value(p.get("Rent_Price")) == amount]

    # Bedrooms (e.g., 2 bhk, 3 bedrooms)
    bhk_match = re.search(r"(\d+)\s*(bhk|bed(room)?s?)", user_input)
    if bhk_match:
        bhk = int(bhk_match.group(1))
        results = [p for p in results if get_numeric_value(p.get("Bedrooms")) == bhk]

    # Area
    for p in data:
        if "area" in user_input and p.get("Area", "").lower() in user_input:
            results = [prop for prop in results if prop.get("Area", "").lower() in user_input]

    # Zone
    for p in data:
        if "zone" in user_input and p.get("Zone", "").lower() in user_input:
            results = [prop for prop in results if prop.get("Zone", "").lower() in user_input]

    # Facilities
    if "parking" in user_input or "car" in user_input:
        results = [p for p in results if p.get("Facilities", {}).get("parking", 0) == 1]
    if "gym" in user_input:
        results = [p for p in results if p.get("Facilities", {}).get("gym", 0) == 1]
    if "swimming" in user_input:
        results = [p for p in results if p.get("Facilities", {}).get("swimming_pool", 0) == 1]

    return results


# --- Format results ---
def format_property(prop):
    return (
        f"üè† ID: {prop.get('property_id', 'N/A')} | Rent: ‚Çπ{prop.get('Rent_Price', 'N/A')} | "
        f"{prop.get('Bedrooms', 'N/A')} BHK | {prop.get('Size_In_Sqft', 'Unknown')} sqft | "
        f"Area: {prop.get('Area', 'N/A')} | Zone: {prop.get('Zone', 'N/A')}"
    )


# --- Chatbot ---
def chatbot_response():
    print("ü§ñ Welcome to Property Chatbot!")
    print("You can ask things like:")
    print("- show me 2 bhk flats in Andheri with rent below 20000")
    print("- find flats in zone west with gym and swimming pool")
    print("- property with parking in Bandra")
    print("Type 'exit' to quit.\n")

    while True:
        user_input = input("You: ").strip().lower()
        if user_input == "exit":
            print("Goodbye üëã")
            break

        results = filter_properties(user_input, properties_data)

        if not results:
            print("‚ùå No properties found matching your query.")
        else:
            print(f"‚úÖ Found {len(results)} property(s):\n")
            for prop in results[:10]:  # Show only first 10
                print(format_property(prop))
            print("-" * 40)


if __name__ == "__main__":
    chatbot_response()
```

```python
import json
import re
from collections import defaultdict

# --- Load properties from JSON file ---
try:
    with open("property_data.json", "r") as f:
        properties_data = json.load(f)
except FileNotFoundError:
    print("Error: 'property_data.json' not found. Please ensure the file exists.")
    properties_data = []
except json.JSONDecodeError:
    print("Error: Could not decode 'property_data.json'. Please check its format.")
    properties_data = []

# --- Helper functions for normalization ---
def normalize_area_name(area_name):
    return str(area_name).replace(" ", "").lower().strip()

def normalize_zone_name(zone_name):
    return str(zone_name).replace(" ", "").lower().strip()

def normalize_facility_name(facility_name):
    return str(facility_name).replace(" ", "_").lower().strip()

def normalize_amenity_name(name):
    return str(name).replace(" ", "_").lower().strip()

def normalize_room_name(room_name):
    return str(room_name).replace(" ", "").lower().strip()

def normalize_property_type_name(prop_type):
    return str(prop_type).lower().strip()

def get_numeric_value(value):
    """Extract integer from strings, supporting both digits and spelled-out numbers.
        Returns None if no number is found."""
    if not value:
        return None
    
    # Dictionary for number words
    num_words = {
        'one': 1, 'two': 2, 'three': 3, 'four': 4, 'five': 5,
        'six': 6, 'seven': 7, 'eight': 8, 'nine': 9, 'ten': 10,
        'eleven': 11, 'twelve': 12, 'thirteen': 13, 'fourteen': 14,
        'fifteen': 15, 'sixteen': 16, 'seventeen': 17, 'eighteen': 18,
        'nineteen': 19, 'twenty': 20, 'thirty': 30, 'forty': 40,
        'fifty': 50, 'sixty': 60, 'seventy': 70, 'eighty': 80,
        'ninety': 90, 'hundred': 100, 'thousand': 1000, 'lakh': 100000,
        'million': 1000000, 'crore': 10000000
    }

    # Normalize the input string
    normalized_value = str(value).lower().strip().replace('-', ' ')
    
    # Check for number words first
    total_value = 0
    current_number = 0
    words = normalized_value.split()
    
    for word in words:
        if word in num_words:
            if num_words[word] >= 100:
                total_value += current_number * num_words[word]
                current_number = 0
            else:
                current_number += num_words[word]
    total_value += current_number

    # If number words were found, return the calculated value
    if total_value > 0:
        return total_value
    
    # If no number words, search for digits
    match = re.search(r"\d+", normalized_value)
    return int(match.group()) if match else None


# --- Dynamically get all unique values from the dataset ---
ALL_AREAS = sorted(
    list(set(normalize_area_name(p.get("Area", "N/A")) for p in properties_data))
)

ALL_ZONES = sorted(
    list(set(normalize_zone_name(p.get("Zone", "N/A")) for p in properties_data))
)

ALL_FACILITIES = []
for p in properties_data:
    if "Facilities" in p and isinstance(p["Facilities"], dict):
        ALL_FACILITIES = sorted(list(p["Facilities"].keys()))
        break

ALL_NEARBY_AMENITIES = []
for p in properties_data:
    if "Nearby_Amenities" in p and isinstance(p["Nearby_Amenities"], dict):
        ALL_NEARBY_AMENITIES = sorted(list(p["Nearby_Amenities"].keys()))
        break

ALL_ROOM_TYPES = sorted(
    list(set(normalize_room_name(p.get("Room_Details", {}).get("Rooms", "N/A"))
             for p in properties_data if p.get("Room_Details", {}).get("Rooms")))
)

ALL_PROPERTY_TYPES = sorted(
    list(set(normalize_property_type_name(p.get("Room_Details", {}).get("Type", "N/A"))
             for p in properties_data if p.get("Room_Details", {}).get("Type")))
)



# --- Filtering logic ---
def filter_properties(user_input, field, data):
    filtered_properties = []

    # Map search field to the actual key in properties_data
    data_field_map = {
        "size": "Size_In_Sqft", "carpet": "Carpet_Area_Sqft", "age": "Property_Age", "brokerage": "Brokerage",
        "furnishing": "Furnishing_Status", "amenities": "Number_Of_Amenities", "security": "Security_Deposite", "rent": "Rent_Price",
        "area": "Area", "zone": "Zone", "bedrooms": "Bedrooms", "bathrooms": "Bathrooms", "balcony": "Balcony",
        "floor_no": "Floor_No", "total_floors": "Total_floors_In_Building", "maintenance": "Maintenance_Charge",
        "recommended_for": "Recommended_For", "water_supply": "Water_Supply_Type", "society_type": "Society_Type",
        "road_connectivity": "Road_Connectivity", "facilities": "Facilities", "nearby_amenities": "Nearby_Amenities",
        "room_type": "Room_Details", "property_type": "Room_Details", "id": "property_id"
    }
    data_field = data_field_map.get(field)
    if not data_field:
        return []

    # --- String fields ---
    normalized_user_input = user_input.lower().strip()
    if field in ["brokerage", "furnishing", "maintenance", "recommended_for", "water_supply", "society_type"]:
        filtered_properties = [p for p in data if str(p.get(data_field, "N/A")).lower() == normalized_user_input]

    elif field == "facilities":
        user_facilities = [normalize_facility_name(f) for f in user_input.split(',')]
        filtered_properties = [
            p for p in data
            if all(
                any(normalize_facility_name(k) == fac and v == 1
                    for k, v in p.get("Facilities", {}).items())
                for fac in user_facilities
            )
        ]


    elif field == "nearby_amenities":
        user_facilities = [normalize_facility_name(f) for f in user_input.split(',')]
        filtered_properties = [
            p for p in data
            if all(
                any(normalize_facility_name(k) == fac and v == 1
                    for k, v in p.get("Nearby_Amenities", {}).items())
                for fac in user_facilities
            )
        ]

    elif field == "room_type":
        filtered_properties = [p for p in data if normalize_room_name(p.get("Room_Details", {}).get("Rooms", "")) == normalized_user_input]

    elif field == "property_type":
        filtered_properties = [p for p in data if normalize_property_type_name(p.get("Room_Details", {}).get("Type", "")) == normalized_user_input]

    elif field == "area":
        filtered_properties = [p for p in data if normalize_area_name(p.get("Area", "N/A")) == normalize_area_name(user_input)]

    elif field == "zone":
        filtered_properties = [p for p in data if normalize_zone_name(p.get("Zone", "N/A")) == normalize_zone_name(user_input)]

    elif field == "id":
        filtered_properties = [p for p in data if str(p.get(data_field)) == user_input.strip()]

    else:
        try:
            val = get_numeric_value(user_input)
    
            if user_input.startswith("below"):
                filtered_properties = [
                    p for p in data
                    if get_numeric_value(p.get(data_field)) is not None
                    and get_numeric_value(p.get(data_field)) < val
                ]
            elif user_input.startswith("above"):
                filtered_properties = [
                    p for p in data
                    if get_numeric_value(p.get(data_field)) is not None
                    and get_numeric_value(p.get(data_field)) > val
                ]
            elif user_input.startswith("between"):
                nums = re.findall(r"\d+", user_input)
                if len(nums) == 2:
                    low, high = int(nums[0]), int(nums[1])
                    filtered_properties = [
                        p for p in data
                        if get_numeric_value(p.get(data_field)) is not None
                        and low <= get_numeric_value(p.get(data_field)) <= high
                    ]
            else:
                filtered_properties = [
                    p for p in data
                    if get_numeric_value(p.get(data_field)) == val
                ]
    
        except Exception:
            return []

    return filtered_properties

# --- Format results ---
def format_property(prop):
    property_id = prop.get('property_id', 'N/A')
    rent_price = prop.get('Rent_Price', 'N/A')
    size = prop.get('Size_In_Sqft', 'Unknown')
    carpet_area = prop.get('Carpet_Area_Sqft', 'Unknown')
    security_deposit = prop.get('Security_Deposite', 'N/A')
    brokerage = prop.get('Brokerage', 'N/A')
    furnishing_status = prop.get('Furnishing_Status', 'N/A')
    amenities = prop.get('Number_Of_Amenities',0)
    age = prop.get('Property_Age', 'Unknown')
    area = prop.get('Area', 'N/A')
    zone = prop.get('Zone', 'N/A')
    bedrooms = prop.get('Bedrooms', 'N/A')
    bathrooms = prop.get('Bathrooms', 'N/A')
    balcony = prop.get('Balcony', 'N/A')
    floor_no = prop.get('Floor_No', 'N/A')
    total_floors = prop.get('Total_floors_In_Building', 'N/A')
    maintenance = prop.get('Maintenance_Charge', 'N/A')
    recommended_for = prop.get('Recommended_For', 'N/A')
    water_supply = prop.get('Water_Supply_Type', 'N/A')
    society_type = prop.get('Society_Type', 'N/A')
    road_connectivity = prop.get('Road_Connectivity', 'N/A')
    facilities_list = [k.replace("_", " ").title() for k, v in prop.get("Facilities", {}).items() if v == 1]
    facilities = ', '.join(facilities_list) if facilities_list else 'None'
    nearby_amenities_list = [k.replace("_", " ").title() for k, v in prop.get("Nearby_Amenities", {}).items() if v == 1]
    nearby_amenities = ', '.join(nearby_amenities_list) if nearby_amenities_list else 'None'
    rooms = prop.get("Room_Details", {}).get("Rooms", "N/A")
    property_type = prop.get("Room_Details", {}).get("Type", "N/A")

    return (
        f"ID: {property_id} | Rent: ‚Çπ{rent_price} | Size: {size} sqft | Carpet Area: {carpet_area} sqft\n"
        f"Rooms: {rooms} | Property Type: {property_type} | Bedrooms: {bedrooms} | Bathrooms: {bathrooms} | Balcony: {balcony}\n"
        f"Furnishing: {furnishing_status} | Security Deposit: ‚Çπ{security_deposit} | Brokerage: {brokerage}\n"
        f"Amenities: {amenities}\n"
        f"Facilities: {facilities}\n"
        f"Nearby Amenities: {nearby_amenities}\n"
        f"Floor: {floor_no}/{total_floors} | Maintenance: {maintenance} | Recommended For: {recommended_for}\n"
        f"Water Supply: {water_supply} | Society: {society_type} | Road Connectivity: {road_connectivity} km\n"
        f"Age: {age} years | Area: {area} | Zone: {zone}\n"
        f"----------------------------------------"
    )

def process_single_line_query(query):
    """Parses a single-line query like 'provide me size under 800 sqft'."""
    operator_map = {
        'under': 'below', 'less than': 'below', 'below': 'below',
        'above': 'above', 'over': 'above', 'more than': 'above',
        'between': 'between'
    }
    
    # Map from verbose names to internal names
    field_map = {
        'size': 'size', 'carpet area': 'carpet', 'age of property': 'age',
        'brokerage': 'brokerage', 'property id': 'id', 'amenities': 'amenities',
        'furnishing status': 'furnishing', 'security deposit': 'security',
        'rent price': 'rent', 'area': 'area', 'zone': 'zone',
        'bedrooms': 'bedrooms', 'bathrooms': 'bathrooms', 'balcony': 'balcony',
        'floor number': 'floor_no', 'total floors': 'total_floors',
        'maintenance charge': 'maintenance', 'recommended for': 'recommended_for',
        'water supply type': 'water_supply', 'society type': 'society_type',
        'road connectivity': 'road_connectivity', 'facilities': 'facilities',
        'nearby amenities': 'nearby_amenities', 'room type': 'room_type',
        'property type': 'property_type'
    }

    # Regex to capture the field, operator, and value.
    # It handles both single-word and multi-word fields like "carpet area"
    # and optional starting phrases like "provide me".
    match = re.search(
        r'(?:provide me|find|show me|search for)?\s*(?:\s*a)?\s*('
        + '|'.join(re.escape(k) for k in field_map.keys())
        + r')\s+('
        + '|'.join(re.escape(k) for k in operator_map.keys())
        + r')\s+(.+)$',
        query, re.IGNORECASE
    )

    if match:
        field_name = field_map.get(match.group(1).lower())
        operator = operator_map.get(match.group(2).lower())
        value = match.group(3).strip()

        # Re-format the value for the filter function
        if operator in ['below', 'above']:
            # The filter function expects "below value" or "above value"
            value = f"{operator} {value}"
        elif operator == 'between':
            # The filter function expects "between val1 and val2"
            # Split the value string to get two numbers
            between_match = re.search(r'(\d+)\s+and\s+(\d+)', value)
            if between_match:
                value = f"between {between_match.group(1)} and {between_match.group(2)}"
        
        return field_name, value
    return None, None

# --- The existing property search chatbot logic, now the 'inner' chatbot ---
def property_search_chatbot():
    print("Welcome to Property Search Chatbot!")
    print("\nSearch Options:")
    print(
        "1. Size (sqft)\n2. Carpet Area (sqft)\n3. Age of Property\n4. Brokerage (yes/no)\n5. Property ID (exact match)\n"
        "6. Amenities\n7. Furnishing Status\n8. Security Deposit\n9. Rent Price\n10. Area\n11. Zone\n"
        "12. Bedrooms\n13. Bathrooms\n14. Balcony\n15. Floor Number\n16. Total Floors\n17. Maintenance Charge\n"
        "18. Recommended For\n19. Water Supply Type\n20. Society Type\n21. Road Connectivity\n22. Facilities\n"
        "23. Nearby Amenities\n24. Room Type (e.g., 1 BHK)\n25. Property Type (e.g., Flat)\n"
    )
    print("üí° You can combine multiple search fields using '+' or a comma.")
    print("Example: 9+10 (Rent + Area) or 'rent, area'")
    print("For numeric values, you can use 'above 1000' or 'above one thousand', 'below 5000', 'between 1000 and 5000'.")
    print("üí° **New!** You can make a full query in one line. For example: 'provide me a size under 800 sqft'.")
    print("Type 'back' to return to the main menu.")

    CATEGORY_OPTIONS = {
    "brokerage": sorted(list(set(str(p.get("Brokerage", "N/A")).lower() for p in properties_data))),
    "furnishing": sorted(list(set(str(p.get("Furnishing_Status", "N/A")).lower() for p in properties_data))),
    "maintenance": sorted(list(set(str(p.get("Maintenance_Charge", "N/A")).lower() for p in properties_data))),
    "recommended_for": sorted(list(set(str(p.get("Recommended_For", "N/A")).lower() for p in properties_data))),
    "water_supply": sorted(list(set(str(p.get("Water_Supply_Type", "N/A")).lower() for p in properties_data))),
    "society_type": sorted(list(set(str(p.get("Society_Type", "N/A")).lower() for p in properties_data))),
    "area": ALL_AREAS,
    "zone": ALL_ZONES,
    "room_type": ALL_ROOM_TYPES,
    "property_type": ALL_PROPERTY_TYPES,
    "facilities": ALL_FACILITIES,
    "nearby_amenities": ALL_NEARBY_AMENITIES
    }

    # Map for verbal and numerical input for search fields
    search_map = {
        "1": "size", "size": "size",
        "2": "carpet", "carpet area": "carpet",
        "3": "age", "age of property": "age",
        "4": "brokerage",
        "5": "id", "property id": "id",
        "6": "amenities",
        "7": "furnishing", "furnishing status": "furnishing",
        "8": "security", "security deposit": "security",
        "9": "rent", "rent price": "rent",
        "10": "area",
        "11": "zone",
        "12": "bedrooms",
        "13": "bathrooms",
        "14": "balcony",
        "15": "floor_no", "floor number": "floor_no",
        "16": "total_floors", "total floors": "total_floors",
        "17": "maintenance", "maintenance charge": "maintenance",
        "18": "recommended_for", "recommended for": "recommended_for",
        "19": "water_supply", "water supply type": "water_supply",
        "20": "society_type", "society type": "society_type",
        "21": "road_connectivity", "road connectivity": "road_connectivity",
        "22": "facilities",
        "23": "nearby_amenities", "nearby amenities": "nearby_amenities",
        "24": "room_type", "room type": "room_type",
        "25": "property_type", "property type": "property_type"
    }

    while True:
        field_input = input("\nEnter your query: ").strip().lower()
        if field_input == "back":
            print("Returning to main menu.")
            break
        
        combined_filters = {}
        
        # Try to parse as a single-line query
        field, value = process_single_line_query(field_input)
        
        if field and value:
            # Single-line query detected and parsed
            combined_filters[field] = value
        else:
            # Fallback to the original multi-step process
            field_tokens = re.split(r'[+,]', field_input)
            
            selected_fields = []
            is_valid_input = True
            for token in field_tokens:
                token = token.strip()
                if token in search_map:
                    selected_fields.append(search_map[token])
                else:
                    print(f"‚ùå Invalid field input: '{token}'. Please use a valid number or name from the list.")
                    is_valid_input = False
                    break
            
            if not is_valid_input:
                continue

            for field in selected_fields:
                if field in CATEGORY_OPTIONS and CATEGORY_OPTIONS[field]:
                    options = CATEGORY_OPTIONS[field]
                    print(f"\nAvailable options for {field.replace('_',' ').title()}:")
                    for idx, option in enumerate(options, start=1):
                        print(f"{idx}. {option}")
                    
                    choice = input(f"Enter {field.replace('_',' ').title()} value (choose number or type value): ").strip().lower()
                    
                    # If user entered a number, map to option
                    if choice.isdigit():
                        choice_idx = int(choice) - 1
                        if 0 <= choice_idx < len(options):
                            user_input = options[choice_idx]
                        else:
                            print("‚ùå Invalid choice, defaulting to your raw input.")
                            user_input = choice
                    else:
                        user_input = choice
                else:
                    user_input = input(f"Enter {field.replace('_',' ').title()} value: ").strip()
                
                combined_filters[field] = user_input


        # Special case for 'id'
        if "id" in combined_filters:
            prop_id = combined_filters["id"]
            prop = next((p for p in properties_data if str(p.get("property_id")) == prop_id), None)
            if prop:
                print("\n‚úÖ Property Found:")
                print(format_property(prop))
            else:
                print("‚ùå Property not found.")
            continue
        
        results = properties_data
        for field, value in combined_filters.items():
            results = filter_properties(value, field, results)
        
        if not results:
            print("‚ùå No properties found matching your combined search.")
        else:
            print(f"\n‚úÖ Found {len(results)} properties matching your combined query.")
            grouped_results = defaultdict(list)
            for prop in results:
                property_type = prop.get("Room_Details", {}).get("Type", "Other/Unspecified Type")
                grouped_results[property_type].append(prop)
            
            for prop_type, props in grouped_results.items():
                print(f"\n--- üè† Property Type: {str(prop_type).title()} ({len(props)} results) ---")
                for prop in props:
                    print(format_property(prop))

# --- Main 'outer' chatbot logic ---
def main_chatbot():
    print("Welcome to the Main Chatbot!")
    print("I can help you with a few things. Here are my current commands:")
    print("1. 'search properties' - Start the property search tool.")
    print("2. 'exit' - End the conversation.")

    while True:
        user_input = input("\nWhat can I help you with? ").strip().lower()

        if user_input == 'search properties':
            # This is where the inner chatbot is called
            property_search_chatbot()
        elif user_input == 'exit':
            print("Goodbye! Thank you for using the chatbot.")
            break
        else:
            print("I'm sorry, I don't understand that command. Please type 'search properties' or 'exit'.")

if __name__ == "__main__":
    main_chatbot()
```
